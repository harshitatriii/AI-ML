# Support Vector Regression (SVR)

## Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset = pd.read_csv("/content/Position_Salaries.csv")
X = dataset.iloc[ : , 1:2]
y = dataset.iloc[ : , 2:3]
```

```
print(X)
```

```
print(y)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_scaled = sc_X.fit_transform(X)

sc_y = StandardScaler()
y_scaled = sc_y.fit_transform(y)
```

## Training the SVR model on the whole dataset

```
from sklearn.svm import SVR
regressor = SVR(kernel="rbf")
regressor.fit(X_scaled , y_scaled)
```

## Predicting a new result

```
input_scaled = sc_X.transform([[6.5]])
print(sc_y.inverse_transform(regressor.predict([[6.5]]).reshape(-1,1)))
print("Estimated Salary for a 6.5 Level employee is : ", sc_y.inverse_transform(regressor.predict(input_scaled).reshape(-1,1)))
# print(regressor.predict([[]]))
```

## Visualising the SVR results

```
plt.scatter(X, y , color = "red")
plt.plot(X , sc_y.inverse_transform(regressor.predict(X_scaled).reshape(-1,1)))
plt.title("Truth or Bluff (SVR Model)")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
